```python
# prompt: generate a code to use causalLM using huggingfae



from transformers import pipeline, set_seed

generator = pipeline('text-generation', model="navaneeth45/Qwen2.5-1.5B-thinking-reasoning-model-V0",device_map = "auto")


```

```python
promtps = ["Three gods are posed before you—one always tells the truth, one always lies, and one answers randomly. They respond using words that mean “yes” and “no,” but you don’t know which word corresponds to which answer. What three yes-or-no questions can you ask to determine which god is which?",
           "There were 27 boys and 35 girls on the playground at recess. There were _____ children on the playground at recess."]
```

```python
prompt = promtps[0]
question = "Problem:" + prompt + "Reasoning:\n"
generated_text = generator(prompt, max_new_tokens=100)
for text in generated_text:
  print(text['generated_text'])
```
